```python
import matplotlib.pyplot as plt

import numpy as np
import shapefile

# Open the existing shapefile
sf = shapefile.Reader("india_st.shf")

# Create a new shapefile writer
w = shapefile.Writer("india_st_copy.shf", shapeType=sf.shapeType)

# Copy the fields from the existing shapefile to the new one
w.fields = sf.fields

# Read and copy the records from the existing shapefile to the new one
records = sf.records()
for record in records:
    w.record(*record)

# Copy the geometries from the existing shapefile to the new one
for shape in sf.shapes():
    w.shape(shape)

# Save and close the new shapefile, which will automatically create the .shx and .dbf files
w.close()



gdf_states = gpd.read_file('india_st.shp')
gdf_states.head()
# gdf_states.plot()

merged = gdf_states.set_index('STATE').join(states.set_index('state'))
merged.head()



# Create a mask to identify states with missing farmer counts
missing_mask = merged['district.farmerCount'].isnull()

# Plot states with missing farmer counts in a separate color or style
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Farmer Count according to state ', fontdict={'fontsize': '15', 'fontweight' : '3'})

# Plot states with non-missing farmer counts
merged[~missing_mask].plot(column='district.farmerCount', cmap='Greys', linewidth=0.5, ax=ax, edgecolor='0.2', legend=True, alpha=0.7)

# Plot states with missing farmer counts
merged[missing_mask].plot(color='White', linewidth=0.5, ax=ax, edgecolor='0.2', hatch='/')

plt.show()
```
